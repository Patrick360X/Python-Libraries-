# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# Functional Way of plotting graph using matplotlib

plt.plot(x,y)

plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Line Graph')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object Oriented Ways of plotting graphs using matplotlib

fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)

axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Graph plotted using Object oriented method')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'g')
axes1.set_title('Outer Plot')

axes2.plot(y,x,'--')
axes2.set_title('Inner Plot')
```

# Matplotlib Intermediate 

```python
fig, axes = plt.subplots(nrows = 1, ncols= 2)

#plt.tight_layout()

# for current_ax in axes:
#     current_ax.plot(x,y)

axes[0].plot(x,y)                            # Axes being iterable objects we can also index them and plot accordingly
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

```python
axes
```

# Figure Size and DPI 

```python
fig = plt.figure(figsize=(6,2), dpi = 110)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1,figsize=(8,4))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('My Plots.png',dpi = 1000)
```

# Creating Legends

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cubed')

ax.legend(loc=(0.1,0.1))
```

# Matplotlib Part -3 

Plot Customisation like line width,color and line types

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color= 'Purple', linewidth = 3, linestyle = '--', marker='o', markersize = '10', markerfacecolor = 'blue') 

#or we can use RGB Hex code like #000000 to mention a specific line color
```

# Plot Controls

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color= 'Purple', linewidth = 3, linestyle = '--') 

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
