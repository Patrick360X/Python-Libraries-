# Matrix Plot

```python
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline

tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')

tips.head()
```

```python
flights.head()
```

```python
tc = tips.corr(numeric_only=True)
```

```python
sns.heatmap(tc,annot=True,cmap='coolwarm')
```

```python
flights
```

```python
fc = flights.corr(numeric_only=True)
```

```python
sns.heatmap(fc)
```

```python
fp =flights.pivot(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(fp,linecolor='white',linewidths=1)
```

```python
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
```

# Grid

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
g = sns.PairGrid(iris)

# g.map(sns.scatterplot)

g.map_diag(sns.histplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
fg = sns.FacetGrid(data=tips,col='time',row='smoker')
fg.map(sns.histplot,'total_bill')
```

# Regression Plots

```python
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex',markers=['o','v'], scatter_kws={'s':50})            # lmplot(): Linear model plot
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day',hue='sex', aspect=0.6)
```
